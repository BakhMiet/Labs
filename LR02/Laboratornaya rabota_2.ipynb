```python

# —Ä–∏—Å—É–µ–º —Å–∞–º—É –ª–∏–Ω–∏—é
x1_range = np.arange(0, 3, 0.1)
plt.plot(x1_range, line_x1(x1_range), color='blue')
x1_range = np.arange(0, 3, 0.1)
plt.plot(x1_range, line_x2(x1_range), color='blue')

# —Ä–∏—Å—É–µ–º —Ç–æ—á–∫–∏
x1x2 = np.random.rand(200, 2)
for x1, x2 in x1x2:
    value_1 = line_1(x1, x2)
    value_2 = line_2(x1, x2)
    # —Å–∫–ª–µ–∏–≤–∞–µ–º "–∑–æ–Ω—ã"
    if value_1 > 0 or value_2 > 0:
        value = 1
    else:
        value = value_1

    if (value == 0): # —Å–∏–Ω–∏–µ ‚Äî –Ω–∞ –ª–∏–Ω–∏–∏
        plt.plot(x1, x2, 'o', color='blue')
    elif (value > 0): # –∑–µ–ª—ë–Ω—ã–µ ‚Äî –≤—ã—à–µ –ª–∏–Ω–∏–∏
        plt.plot(x1, x2, 'o', color='green')
    elif (value < 0): # –∫—Ä–∞—Å–Ω—ã–µ ‚Äî –Ω–∏–∂–µ –ª–∏–Ω–∏–∏
        plt.plot(x1, x2, 'o', color='red')

# –ø—Ä–æ—Å—Ç–∞–≤–ª—è–µ–º –Ω–∞–∑–≤–∞–Ω–∏—è –æ—Å–µ–π
plt.xlabel('x1')
plt.ylabel('x2')
plt.xlim([0, 1])
plt.ylim([0, 1])
# –Ω–∞ —ç–∫—Ä–∞–Ω!
plt.show()
print("–†–∏—Å—É–Ω–æ–∫ 2. –†–∞–∑–¥–µ–ª–µ–Ω–∏–µ –∫–ª–∞—Å—Å–æ–≤ –¥–≤—É–º—è –ø–ª–æ—Å–∫–æ—Å—Ç—è–º–∏")
#–í—Ç–æ—Ä–∞—è –∏—Ç–µ—Ä–∞—Ü–∏—è –∞–ª–≥–æ—Ä–∏—Ç–º–∞:

# —Ä–∏—Å—É–µ–º —Å–∞–º—É –ª–∏–Ω–∏—é
x1_range = np.arange(0, 1, 0.1)
plt.plot(x1_range, line_x3(x1_range), color='blue')

# —Ä–∏—Å—É–µ–º —Ç–æ—á–∫–∏
x1x2 = np.random.rand(200, 2)
for x1, x2 in x1x2:
    value = line_3(x1, x2)
    if (value == 0): # —Å–∏–Ω–∏–µ ‚Äî –Ω–∞ –ª–∏–Ω–∏–∏
        plt.plot(x1, x2, 'o', color='blue')
    elif (value > 0): # –∑–µ–ª—ë–Ω—ã–µ ‚Äî –≤—ã—à–µ –ª–∏–Ω–∏–∏
        plt.plot(x1, x2, 'o', color='green')
    elif (value < 0): # –∫—Ä–∞—Å–Ω—ã–µ ‚Äî –Ω–∏–∂–µ –ª–∏–Ω–∏–∏
        plt.plot(x1, x2, 'o', color='red')

# –ø—Ä–æ—Å—Ç–∞–≤–ª—è–µ–º –Ω–∞–∑–≤–∞–Ω–∏—è –æ—Å–µ–π
plt.xlabel('x1')
plt.ylabel('x2')
plt.xlim([0, 1])
plt.ylim([0, 1])
# –Ω–∞ —ç–∫—Ä–∞–Ω!
plt.show()
print("–†–∏—Å—É–Ω–æ–∫ 3. –°–∫–ª–µ–π–∫–∞ –ø–ª–æ—Å–∫–æ—Å—Ç–µ–π –∏ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –≤—ã—Ö–æ–¥–∞")
```

```python
vals = ((0, 0),(0, 1),(1, 0),(1, 1))
for val in vals:
    res1 = 0.5 - val[0] + val[1] # –Ω–µ–π—Ä–æ–Ω —Å–∫—Ä—ã—Ç—ã–π 1
    if res1 > 0.5:
        res1 = 1
    else:
        res1 = 0

    res2 = 0.5 + val[0] - val[1] # –Ω–µ–π—Ä–æ–Ω —Å–∫—Ä—ã—Ç—ã–π 2
    if res2 > 0.5:
        res2 = 1
    else:
        res2 = 0

    res3 = 0.5 + res1 + res2    # –Ω–µ–π—Ä–æ–Ω –≤—ã—Ö–æ–¥–Ω–æ–π
    if res3 > 0.5:
        res3 = 1
    else:
        res3 = 0
    print(val, ':  ', res3)
```

–í –¥–∞–Ω–Ω–æ–π —Ä–∞–±–æ—Ç–µ —ç—Ç–æ–π –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏ –¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ. –î–æ–±–∞–≤–∏–º, —á—Ç–æ –æ–±—â–∞—è —Å—É—Ç—å –∞–ª–≥–æ—Ä–∏—Ç–º–∞ - –Ω–∞—Ö–æ–∂–¥–µ–Ω–∏–µ –≤–∑–∞–∏–º–æ—Å–≤—è–∑–∏ –≤—ã—Ö–æ–¥–Ω–æ–≥–æ —Å–∏–≥–Ω–∞–ª–∞ Y –∏ –≤—ã—Ö–æ–¥–æ–≤ –Ω–µ–π—Ä–æ–Ω–æ–≤ –ø—Ä–µ–¥—ã–¥—É—â–∏—Ö —Å–ª–æ–µ–≤ –¥–ª—è –ø–æ—Å–ª–µ–¥—É—é—â–µ–≥–æ –Ω–∞—Ö–æ–∂–¥–µ–Ω–∏—è –≥–ª–æ–±–∞–ª—å–Ω–æ–≥–æ (–∏–ª–∏, –ø–æ—Ä–æ–π, –ª–æ–∫–∞–ª—å–Ω–æ–≥–æ) –º–∏–Ω–∏–º—É–º–∞ –æ—à–∏–±–∫–∏ –ø—É—Ç–µ–º "–ø—å—è–Ω–æ–≥–æ —à–∞—Ç–∞–Ω–∏—è" –æ—Ç–≤–µ—Ç–∞ (–º–æ–¥–µ–ª–∏) –≤ –µ–≥–æ —Å—Ç–æ—Ä–æ–Ω—É.

**–ò–º–ø–æ—Ä—Ç–∏—Ä—É–µ–º –Ω—É–∂–Ω—ã–µ –±–∏–±–ª–∏–æ—Ç–µ–∫–∏**

–û–ø—Ä–µ–¥–µ–ª—è–µ–º —Ñ—É–Ω–∫—Ü–∏—é mse_loss, –∫–æ—Ç–æ—Ä–∞—è –≤—ã—á–∏—Å–ª—è–µ—Ç —Å—Ä–µ–¥–Ω–µ–∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω—É—é –æ—à–∏–±–∫—É **(MSE - Mean Squared Error)** –º–µ–∂–¥—É –∏—Å—Ç–∏–Ω–Ω—ã–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ y_true –∏ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ y_pred.

—Ä–µ–¥–Ω–µ–∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–∞—è –æ—à–∏–±–∫–∞ (MSE) ‚Äî —ç—Ç–æ –æ–¥–Ω–∞ –∏–∑ –Ω–∞–∏–±–æ–ª–µ–µ —Ä–∞—Å–ø—Ä–æ—Å—Ç—Ä–∞–Ω–µ–Ω–Ω—ã—Ö —Ñ—É–Ω–∫—Ü–∏–π –ø–æ—Ç–µ—Ä—å, –∏—Å–ø–æ–ª—å–∑—É–µ–º—ã—Ö –≤ –∑–∞–¥–∞—á–∞—Ö —Ä–µ–≥—Ä–µ—Å—Å–∏–∏. –û–Ω–∞ –∏–∑–º–µ—Ä—è–µ—Ç —Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –∫–≤–∞–¥—Ä–∞—Ç–æ–≤ —Ä–∞–∑–Ω–æ—Å—Ç–µ–π –º–µ–∂–¥—É –∏—Å—Ç–∏–Ω–Ω—ã–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ (ùë¶_true) –∏ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ (ùë¶pred) –º–æ–¥–µ–ª–∏.
–§–æ—Ä–º—É–ª–∞ MSE:

\begin{equation}
MSE = \frac{1}{n} \sum_{i=1}^{n} (y_{\text{true}, i} - y_{\text{pred}, i})^2
\end{equation}



**–ì–¥–µ:**
\
\begin{aligned}
    & n - \text{–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω–∞–±–ª—é–¥–µ–Ω–∏–π (–ø—Ä–∏–º–µ—Ä–æ–≤ –≤ –Ω–∞–±–æ—Ä–µ –¥–∞–Ω–Ω—ã—Ö)}, \\
    & y_{\text{true}, i} - \text{–∏—Å—Ç–∏–Ω–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —Ü–µ–ª–µ–≤–æ–π –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π}, \\
    & y_{\text{pred}, i} - \text{–ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ}, \\
    & (y_{\text{true}, i} - y_{\text{pred}, i})^2 - \text{–∫–≤–∞–¥—Ä–∞—Ç –æ—à–∏–±–∫–∏ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –ø—Ä–∏–º–µ—Ä–∞}.
\end{aligned}




```python
def mse_loss(y_true, y_pred): # –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ –æ—à–∏–±–∫–∏
    return 0.5*((y_true - y_pred) ** 2).mean()
```

—Å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ–º NumPy –ø–æ–¥–≥–æ—Ç–∞–≤–ª–∏–≤–∞–µ–º —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω—ã–µ –∏ —Ç–µ—Å—Ç–æ–≤—ã–µ –¥–∞–Ω–Ω—ã–µ –¥–ª—è –∑–∞–¥–∞—á–∏ —Ä–µ–≥—Ä–µ—Å—Å–∏–∏, –ø—Ä–∏–±–ª–∏–∂–∞—è —Ñ—É–Ω–∫—Ü–∏—é —Å–∏–Ω—É—Å–∞.

```python
X_train = np.arange(0, np.pi, 0.2)  # X –æ—Ç 0 –¥–æ œÄ —Å —à–∞–≥–æ–º 0.2
mu, sigma = 0, 0.01  # –°—Ä–µ–¥–Ω–µ–µ –∏ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ —à—É–º–∞
s = np.random.normal(mu, sigma, len(X_train))  # –®—É–º, –¥–æ–±–∞–≤–ª–µ–Ω–Ω—ã–π –∫ y
y_train = np.sin(X_train) + s  # –¶–µ–ª–µ–≤—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è (—Å —à—É–º–æ–º)
X_test = np.arange(0.1,np.pi,0.2)
y_test = np.sin(X_test)

```

–≤—ã–ø–æ–ª–Ω—è–µ–º –º–∞—Å—à—Ç–∞–±–∏—Ä–æ–≤–∞–Ω–∏–µ –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö (X_train –∏ X_test),

–∏—Å–ø–æ–ª—å–∑—É—è StandardScaler –∏–∑ sklearn.preprocessing.
 StandardScaler() ‚Äî —ç—Ç–æ –º–µ—Ç–æ–¥ –Ω–æ—Ä–º–∞–ª–∏–∑–∞—Ü–∏–∏ –¥–∞–Ω–Ω—ã—Ö, –∫–æ—Ç–æ—Ä—ã–π –ø—Ä–∏–≤–æ–¥–∏—Ç –∏—Ö –∫ –Ω—É–ª–µ–≤–æ–º—É —Å—Ä–µ–¥–Ω–µ–º—É –∏ –µ–¥–∏–Ω–∏—á–Ω–æ–º—É —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–º—É –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—é –ø–æ —Ñ–æ—Ä–º—É–ª–µ:

 \begin{equation}
X_{\text{scaled}} = \frac{X - \mu}{\sigma}
\end{equation}


\begin{aligned}
    & \mu - \text{—Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –æ–±—É—á–∞—é—â–∏—Ö –¥–∞–Ω–Ω—ã—Ö}, \\
    & \sigma - \text{—Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ}.
\end{aligned}


```python

sc=StandardScaler() # –ü—Ä–∏–≤–æ–¥–∏–º –¥–∞–Ω–Ω—ã–µ –≤ –Ω—É–∂–Ω—ã–π –¥–ª—è –º–æ–¥–µ–ª–∏ –≤–∏–¥ + –Ω–æ—Ä–º–∏—Ä—É–µ–º
scaler = sc.fit(X_train.reshape(-1, 1))
trainX_scaled = scaler.transform(X_train.reshape(-1, 1))
testX_scaled = scaler.transform(X_test.reshape(-1, 1))

```

–ú–Ω–æ–≥–æ—Å–ª–æ–π–Ω—ã–π –ø–µ—Ä—Ü–µ–ø—Ç—Ä–æ–Ω (MLP) —Ç—Ä–µ–Ω–∏—Ä—É–µ—Ç—Å—è —Å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ–º:
- **–°—Ç–æ—Ö–∞—Å—Ç–∏—á–µ—Å–∫–æ–≥–æ –≥—Ä–∞–¥–∏–µ–Ω—Ç–Ω–æ–≥–æ —Å–ø—É—Å–∫–∞ (SGD)**
- **–ê–¥–∞–º (Adam)** ‚Äî –æ–ø—Ç–∏–º–∏–∑–∏—Ä–æ–≤–∞–Ω–Ω—ã–π —Å—Ç–æ—Ö–∞—Å—Ç–∏—á–µ—Å–∫–∏–π –≥—Ä–∞–¥–∏–µ–Ω—Ç–Ω—ã–π —Å–ø—É—Å–∫
- **L-BFGS (lbfgs)** ‚Äî –º–µ—Ç–æ–¥ —Å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ–º –º–∞—Ç—Ä–∏—Ü—ã –ì–µ—Å—Å–µ

**–ù–µ–π—Ä–æ–Ω–Ω–∞—è —Å–µ—Ç—å**
–ü–æ —É–º–æ–ª—á–∞–Ω–∏—é –∏—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è —Ñ—É–Ω–∫—Ü–∏—è –∞–∫—Ç–∏–≤–∞—Ü–∏–∏ **ReLU**

–í–æ–∑–º–æ–∂–Ω—ã–µ —Ñ—É–Ω–∫—Ü–∏–∏ –∞–∫—Ç–∏–≤–∞—Ü–∏–∏ –¥–ª—è —Å–∫—Ä—ã—Ç—ã—Ö —Å–ª–æ–µ–≤:
-**‚Äòidentity‚Äô** ‚Äî –Ω–µ–π—Ç—Ä–∞–ª—å–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è, –ø–æ–ª–µ–∑–Ω–∞ –¥–ª—è –ª–∏–Ω–µ–π–Ω–æ–≥–æ —Å–∂–∏–∂–µ–Ω–∏—è, f(x) = x
- **‚Äòlogistic‚Äô** ‚Äî –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–∞—è —Å–∏–≥–º–æ–∏–¥–∞, f(x) = 1 / (1 + exp(-x))
- **‚Äòtanh‚Äô** ‚Äî –≥–∏–ø–µ—Ä–±–æ–ª–∏—á–µ—Å–∫–∏–π —Ç–∞–Ω–≥–µ–Ω—Å, f(x) = tanh(x)
- **‚Äòrelu‚Äô** ‚Äî ReLU, f(x) = max(0, x)


```python
regr = MLPRegressor(activation='relu',random_state=1, max_iter=3000, hidden_layer_sizes=(10,2), solver='lbfgs').fit(trainX_scaled, y_train)


preds = [] # —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã —Ä–∞–±–æ—Ç—ã –Ω–µ–π—Ä–æ—Å–µ—Ç–∏
for x in testX_scaled:
    preds.append(regr.predict(x.reshape(-1, 1)))
loss = mse_loss(y_test, preds)
print('loss is', loss)
```

```python
plt.plot(X_test, y_test, label='original')
plt.plot(X_test, preds, label='predicted')
plt.legend()
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
print("–†–∏—Å—É–Ω–æ–∫ 4. –†–µ–∑—É–ª—å—Ç–∞—Ç—ã —Ä–∞–±–æ—Ç—ã —Å–∞–º–æ–æ–±—É—á–∞—é—â–µ–π—Å—è –Ω–µ–π—Ä–Ω–æ–Ω–Ω–æ–π —Å–µ—Ç–∏")
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import cm
plt.style.use('_mpl-gallery')
```

```python
# –°–æ–∑–¥–∞–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö
X = np.arange(-5, 5, 0.25)  # –°–æ–∑–¥–∞–Ω–∏–µ –º–∞—Å—Å–∏–≤–∞ X –æ—Ç -5 –¥–æ 5 —Å —à–∞–≥–æ–º 0.25
Y = np.arange(-5, 5, 0.25)  # –°–æ–∑–¥–∞–Ω–∏–µ –º–∞—Å—Å–∏–≤–∞ Y –æ—Ç -5 –¥–æ 5 —Å —à–∞–≥–æ–º 0.25
X, Y = np.meshgrid(X, Y)  # –°–æ–∑–¥–∞–Ω–∏–µ —Å–µ—Ç–∫–∏ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç –¥–ª—è X –∏ Y

R = np.sqrt(X**2 + Y**2)  # –†–∞—Å—Å—Ç–æ—è–Ω–∏–µ –æ—Ç —Ü–µ–Ω—Ç—Ä–∞ (–≤ —Ä–∞–¥–∏—É—Å–∞—Ö) –¥–ª—è –∫–∞–∂–¥–æ–π —Ç–æ—á–∫–∏
Z = np.sin(R)  # –ó–Ω–∞—á–µ–Ω–∏—è —Ñ—É–Ω–∫—Ü–∏–∏ sin –¥–ª—è –∫–∞–∂–¥–æ–≥–æ —Ä–∞–¥–∏—É—Å–∞
```

```python
# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –ø–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç–∏
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})  # –°–æ–∑–¥–∞–Ω–∏–µ 3D-–ø–æ–¥–≥—Ä–∞—Ñ–∏–∫–∞
ax.plot_surface(X, Y, Z, vmin=Z.min() * 2, cmap=cm.Blues)  # –û—Ç–æ–±—Ä–∞–∂–µ–Ω–∏–µ –ø–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç–∏

# –û—Ç–∫–ª—é—á–µ–Ω–∏–µ –ø–æ–¥–ø–∏—Å–µ–π –æ—Å–µ–π
ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])

plt.show()  # –ü–æ–∫–∞–∑–∞—Ç—å –≥—Ä–∞—Ñ–∏–∫
```
